```python
# 6
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
ser1[~ser1.isin(ser2)]
```

```python
# 7
union = np.union1d(ser1, ser2)
intersection = np.intersect1d(ser1, ser2)
sym_diff = union[~ np.isin(union, intersection)]
sym_diff
```
